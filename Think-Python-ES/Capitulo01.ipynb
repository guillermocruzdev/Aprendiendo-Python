## Operadores aritm√©ticos

Un **operador aritm√©tico** es un s√≠mbolo que representa una operaci√≥n matem√°tica. Por ejemplo, el signo m√°s, `+`, realiza una suma.


````python
30 + 12
````

El signo menos, `-`, es el operador que realiza la resta.


````python
43 - 1
````

El asterisco, `*`, realiza la multiplicaci√≥n.


````python
6 * 7
````

Y la barra diagonal, `/`, realiza la divisi√≥n:


````python
84 / 2
````

Observa que el resultado de la divisi√≥n es `42.0` en lugar de `42`. Esto se debe a que en Python existen dos tipos de n√∫meros:

* **enteros (integers)**, que representan n√∫meros sin parte fraccionaria o decimal, y  

* **n√∫meros de punto flotante (floating-point numbers)**, que representan enteros y n√∫meros con punto decimal.

Si sumas, restas o multiplicas dos enteros, el resultado ser√° un entero.  
Pero si divides dos enteros, el resultado ser√° un n√∫mero de punto flotante.  
Python proporciona otro operador, `//`, que realiza una **divisi√≥n entera**.  
El resultado de la divisi√≥n entera siempre es un n√∫mero entero.


````python
84 // 2
````

La divisi√≥n entera tambi√©n se llama **"divisi√≥n por piso"** (*floor division*) porque siempre redondea hacia abajo (hacia el "piso").


````python
85 // 2
````

Finalmente, el operador `**` realiza la **exponenciaci√≥n**; es decir, eleva un n√∫mero a una potencia:


````python
7 ** 2
````

En algunos otros lenguajes, el acento circunflejo, `^`, se usa para la exponenciaci√≥n, pero en Python  
es un operador a nivel de bits llamado **XOR**.  
Si no est√°s familiarizado con los operadores a nivel de bits, el resultado podr√≠a ser inesperado:


````python
7 ^ 2
````

## Lenguajes formales y naturales

Los **lenguajes naturales** son los lenguajes que hablan las personas, como el ingl√©s, el espa√±ol y el franc√©s. No fueron dise√±ados por personas; evolucionaron de manera natural.

Los **lenguajes formales** son lenguajes dise√±ados por personas para aplicaciones espec√≠ficas.  
Por ejemplo, la notaci√≥n que usan los matem√°ticos es un lenguaje formal que es especialmente bueno para expresar relaciones entre n√∫meros y s√≠mbolos.  
De manera similar, los lenguajes de programaci√≥n son lenguajes formales que han sido dise√±ados para expresar c√≥mputos.


Aunque los lenguajes formales y naturales tienen algunas caracter√≠sticas en com√∫n,  
existen diferencias importantes:

* **Ambig√ºedad:** Los lenguajes naturales est√°n llenos de ambig√ºedad, la cual las personas manejan usando pistas contextuales y otra informaci√≥n.  
  Los lenguajes formales est√°n dise√±ados para ser casi o completamente no ambiguos, lo que significa que cualquier programa tiene exactamente un significado, sin importar el contexto.

* **Redundancia:** Para compensar la ambig√ºedad y reducir malentendidos, los lenguajes naturales usan redundancia.  
  Como resultado, a menudo son verbosos. Los lenguajes formales son menos redundantes y m√°s concisos.

* **Literalidad:** Los lenguajes naturales est√°n llenos de expresiones idiom√°ticas y met√°foras.  
  Los lenguajes formales significan exactamente lo que dicen.


Debido a que todos crecemos hablando lenguajes naturales, a veces es dif√≠cil adaptarse a los lenguajes formales.  
Los lenguajes formales son m√°s densos que los lenguajes naturales, por lo que toma m√°s tiempo leerlos.  
Adem√°s, la estructura es importante, as√≠ que no siempre es mejor leer de arriba abajo o de izquierda a derecha.  
Finalmente, los detalles importan. Peque√±os errores en la ortograf√≠a y la puntuaci√≥n, con los que puedes salirte con la tuya en los lenguajes naturales, pueden hacer una gran diferencia en un lenguaje formal.

### ¬øQu√© es el operador m√≥dulo en Python?

### ¬øPython redondea hacia arriba o hacia abajo?

En Python, la funci√≥n **`round()`** utiliza una estrategia llamada **‚Äúround half to even‚Äù** o **redondeo al par**.

#### üîπ ¬øQu√© significa esto?

Cuando el n√∫mero est√° exactamente a la mitad (por ejemplo: 1.5, 2.5, 3.5), Python **redondea hacia el n√∫mero par m√°s cercano**, no siempre hacia arriba.

#### üîπ Reglas de redondeo

- Si la parte decimal es **< 0.5** ‚Üí redondea **hacia abajo**.
- Si la parte decimal es **> 0.5** ‚Üí redondea **hacia arriba**.
- Si la parte decimal es **exactamente 0.5** ‚Üí redondea al **entero par** m√°s cercano.

#### üîπ Ejemplos

```python
round(2.4)   # 2  -> redondea hacia abajo
round(2.6)   # 3  -> redondea hacia arriba
round(2.5)   # 2  -> 2 es par
round(3.5)   # 4  -> 4 es par
```

#### üîπ ¬øPor qu√© funciona as√≠?

Esta t√©cnica reduce el sesgo estad√≠stico cuando se realizan muchos redondeos,
especialmente en c√°lculos financieros o cient√≠ficos.
