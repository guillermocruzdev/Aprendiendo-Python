## Operadores aritméticos

Un **operador aritmético** es un símbolo que representa una operación matemática. Por ejemplo, el signo más, `+`, realiza una suma.


```python
30 + 12
```

El signo menos, `-`, es el operador que realiza la resta.


```python
43 - 1
```

El asterisco, `*`, realiza la multiplicación.


```python
6 * 7
```

Y la barra diagonal, `/`, realiza la división:


```python
84 / 2
```

Observa que el resultado de la división es `42.0` en lugar de `42`. Esto se debe a que en Python existen dos tipos de números:

* **enteros (integers)**, que representan números sin parte fraccionaria o decimal, y  

* **números de punto flotante (floating-point numbers)**, que representan enteros y números con punto decimal.

Si sumas, restas o multiplicas dos enteros, el resultado será un entero.  
Pero si divides dos enteros, el resultado será un número de punto flotante.  
Python proporciona otro operador, `//`, que realiza una **división entera**.  
El resultado de la división entera siempre es un número entero.


```python
84 // 2
```

La división entera también se llama **"división por piso"** (*floor division*) porque siempre redondea hacia abajo (hacia el "piso").


```python
85 // 2
```

Finalmente, el operador `**` realiza la **exponenciación**; es decir, eleva un número a una potencia:


```python
7 ** 2
```

En algunos otros lenguajes, el acento circunflejo, `^`, se usa para la exponenciación, pero en Python  
es un operador a nivel de bits llamado **XOR**.  
Si no estás familiarizado con los operadores a nivel de bits, el resultado podría ser inesperado:


```python
7 ^ 2
```

No abordaré los operadores a nivel de bits en este libro, pero puedes leer sobre ellos en <http://wiki.python.org/moin/BitwiseOperators>.


## Expresiones

Una colección de operadores y números se llama una **expresión**.  
Una expresión puede contener cualquier cantidad de operadores y números.  
Por ejemplo, aquí hay una expresión que contiene dos operadores.


```python
6 + 6 ** 2
```

Observa que la exponenciación ocurre antes que la suma.  
Python sigue el orden de operaciones que quizá aprendiste en una clase de matemáticas: la exponenciación ocurre antes que la multiplicación y la división, las cuales ocurren antes que la suma y la resta.

En el siguiente ejemplo, la multiplicación ocurre antes que la suma.


```python
12 + 5 * 6
```

Si quieres que la suma ocurra primero, puedes usar paréntesis.


```python
(12 + 5) * 6
```

Cada expresión tiene un **valor**.  
Por ejemplo, la expresión `6 * 7` tiene el valor `42`.


## Funciones aritméticas

Además de los operadores aritméticos, Python proporciona algunas **funciones** que trabajan con números.  
Por ejemplo, la función `round` toma un número de punto flotante y lo redondea al entero más cercano.


```python
round(42.4)
```

```python
round(42.6)
```

La función `abs` calcula el valor absoluto de un número.  
Para un número positivo, el valor absoluto es el mismo número.

```python
abs(42)
```

Para un número negativo, el valor absoluto es positivo.


```python
abs(-42)
```
